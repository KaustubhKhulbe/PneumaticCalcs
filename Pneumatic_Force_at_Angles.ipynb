```
import math
import numpy as np
import matplotlib.pyplot as plt
```

**Input parameters**

Cc - Center to center distance
Fp - Force pneumatic applies
r - radius of pneumatic attachment
A - angle between x-axis and pneumatic end
min_angle - four bar starting angle
max_angle - four bar ending angle

```
Cc = 20
Fp = 30
r = 10
A = 20 #in degrees
min_angle = 0
max_angle = 50
```

Conversions into radians

```
A = A * math.pi / 180
print(A)
```

> Indented block



Calculates sin_a at an angle

```
sin_a = (Cc * math.sin(A)) / (math.sqrt(r**2 + Cc**2 - 2*r*Cc*math.cos(A)))
print(sin_a)
```

Calculates Tp at an angle

```
Tp = Fp * r * sin_a
print(Tp)
```

Function to calculate Tp based on theta

```
def getTp(theta):
  theta = theta * math.pi / 180
  sin_t = (Cc * math.sin(theta)) / (math.sqrt(r**2 + Cc**2 - 2*r*Cc*math.cos(theta)))
  T = Fp * r * sin_t
  return T
```

Function to calculate pneumatic length given theta

```
def getPneumaticLength(theta):
  theta = theta * math.pi / 180
  length = math.sqrt(Cc**2 + r**2 - 2*r*Cc*math.cos(theta))
  return length

  #Cc^2 + r^2 - 2*r*Cc*cos*theta = length^2
```

Graphs the function relative to pneumatic length and also from min_angle to max_angle

```
values = []
values2 = []
values3 = []
for i in range(min_angle, max_angle):
  values.append(getTp(i))
  values2.append(getPneumaticLength(i))
  values3.append(i)

plt.scatter(values2, values)
```
